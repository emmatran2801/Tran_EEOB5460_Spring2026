```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from Bio import SeqIO
from Bio.Seq import Seq
from Bio.Data import CodonTable
from Bio.SeqUtils.ProtParam import ProteinAnalysis
```

```python
os.getcwd()
```

```python
def get_sequences_from_file(fasta_fn):
    sequence_data_dict = {}
    
    for record in SeqIO.parse(fasta_fn, "fasta"):
        description = record.description.split()
        species_name = description[1] + " " + description[2]
        sequence_data_dict[species_name] = record.seq
        
    return(sequence_data_dict)
```

```python
fasta_file = "penguins_cytb.fasta"
sequence_data = get_sequences_from_file(fasta_file)
print(sequence_data)
```

```python
# manual translate function:
def translate_function(string_nucleotides):
    mito_table = CodonTable.unambiguous_dna_by_name["Vertebrate Mitochondrial"]
    seq = string_nucleotides.upper()
    aa_seq = ""
    
    # loop through sequence in steps of 3
    for i in range(0, len(seq) - 2, 3):
        codon = seq[i:i+3]
        # skip incomplete codons
        if len(codon) < 3:
            break
        # stop codon handling
        if codon in mito_table.stop_codons:
            break  # do not include stop in output
        # translate codon → amino acid
        aa = mito_table.forward_table.get(codon, "X")  # "X" for unknown codons
        aa_seq += aa
        
    return aa_seq
```

```python
protein_data_v0 = {}

for species, sequence in sequence_data.items():
    protein_data_v0[species] = translate_function(sequence)
    
print(protein_data_v0)
```

```python
# Biopython translate function:
def translate_biopython(string_nucleotides):
    seq = Seq(string_nucleotides)
    # to_stop=True → stops at first stop codon and excludes it
    protein = seq.translate(table="Vertebrate Mitochondrial", to_stop=True)
    return str(protein)
```

```python
protein_data_v1 = {}

for species, sequence in sequence_data.items():
    protein_data_v1[species] = translate_biopython(sequence)
    
print(protein_data_v1 == protein_data_v0)
```

```python
def compute_molecular_weight(aa_seq):
    # ensure string and clean sequence
    seq = str(aa_seq).replace("*", "").upper()
    # create ProteinAnalysis object
    analysis = ProteinAnalysis(seq)
    # compute molecular weight
    mw = analysis.molecular_weight()
    return mw
```

```python
molecular_weight_data = {}

for species, protein in protein_data_v1.items():
    molecular_weight_data[species] = compute_molecular_weight(protein)
    
print(molecular_weight_data)
```

```python
def compute_gc_content(dna_seq):
    # convert to string and uppercase
    seq = str(dna_seq).upper()
    # avoid division by zero
    if len(seq) == 0:
        return 0
    # count G and C
    gc_count = seq.count("G") + seq.count("C")
    # compute proportion
    gc_content = gc_count / len(seq)
    return gc_content
```

```python
GC_content = {}

for species, sequence in sequence_data.items():
    GC_content[species] = compute_gc_content(sequence)
    
print(GC_content)
```

```python
#%%%%%%%%%%%%%%#
###   MAIN   ###
#%%%%%%%%%%%%%%#

cytb_seqs = get_sequences_from_file("penguins_cytb.fasta") 

penguins_df = pd.read_csv("penguins_mass.csv")
species_list = list(penguins_df.species)

## Add empty columns:
penguins_df["molecular_weight"] = np.nan
penguins_df["gc_content"] = np.nan

## Loop, translate, compute, and fill DataFrame:
for species, dna_seq in cytb_seqs.items():
    # translate DNA → protein
    aa_seq = translate_biopython(dna_seq)
    # compute values
    mw = compute_molecular_weight(aa_seq)
    gc = compute_gc_content(dna_seq)
    # fill DataFrame (match by species)
    penguins_df.loc[penguins_df["species"] == species, "molecular_weight"] = mw
    penguins_df.loc[penguins_df["species"] == species, "gc_content"] = gc
```

```python
## Bar chart of body mass:
sorted_df = penguins_df.sort_values(by="mass", ascending=True)
plt.figure()
plt.barh(sorted_df["species"], sorted_df["mass"])
plt.ylabel("Species")
plt.xlabel("Body Mass (g)")
plt.tight_layout()
plt.show()

print("Q1. Smallest species:", penguins_df.loc[penguins_df["mass"].idxmin(), "species"])
print("Q2. Geographic range: the coasts of New Zealand and southern Australia")
```

```python
## Scatter plot: MW vs GC content:
plt.figure()
plt.scatter(penguins_df["gc_content"], penguins_df["molecular_weight"])
plt.xlabel("GC Content")
plt.ylabel("Molecular Weight")
plt.show()
```

```python
## Save DataFrame:
penguins_df.to_csv("penguins_mass_cytb.csv", index=False)
```

```python
## Bonus - Boxplot of GC content grouped by species
ax = penguins_df.boxplot(column="gc_content", by="species", vert=False)
ax.grid(False)
plt.xlabel("GC Content")
plt.ylabel("Species")
plt.suptitle("")
plt.title("")
plt.show()
```
